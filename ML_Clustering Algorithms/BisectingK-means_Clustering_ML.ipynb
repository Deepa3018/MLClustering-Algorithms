```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import BisectingKMeans
bkmeans = BisectingKMeans(n_clusters=5, random_state=0)
y_bkmeans = bkmeans.fit_predict(X)
```

```python
y_bkmeans
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_bkmeans
```

```python
supervised
```

```python
supervised.to_csv("bisectKmeanscluster.csv",index=False)
```
