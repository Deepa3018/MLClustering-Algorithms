```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import MeanShift
MS=MeanShift(bandwidth=20, seeds=None, bin_seeding=False, min_bin_freq=5, cluster_all=True, n_jobs=None, max_iter=300)
y_MS=MS.fit_predict(X)
```

```python
y_MS
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_MS
```

```python
supervised
```

```python
supervised.to_csv("meanshiftcluster.csv",index=False)
```
