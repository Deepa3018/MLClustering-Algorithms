```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import OPTICS
OP=OPTICS(min_samples=15,metric='minkowski', p=2, cluster_method='xi', eps=None, xi=0.05, predecessor_correction=True, min_cluster_size=5, algorithm='auto', leaf_size=30)
y_OP=OP.fit_predict(X)
```

```python
y_OP
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_OP
```

```python
supervised
```

```python
supervised.to_csv("opticscluster.csv",index=False)
```
