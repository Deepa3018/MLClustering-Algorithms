```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
SC=SpectralClustering(n_clusters=5,assign_labels='discretize', random_state=0)
y_SC=SC.fit_predict(X)
```

```python
y_SC
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_SC
```

```python
supervised
```

```python
supervised.to_csv("spectralcluster.csv",index=False)
```
